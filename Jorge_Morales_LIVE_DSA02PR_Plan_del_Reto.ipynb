```python
import pandas as pd
import numpy as np
from IPython.display import display, HTML
```

```python
df = pd.read_csv('/content/Employee Attrition.csv')
#only the 5 first rows are printed to be familiar with the dataframe we are going to work with, but the shape of the dataframe is displayed at the beginning
print(f"El dataframe cuenta con la siguiente forma {df.shape}")
df.head(5)
```

5. Obten la siguiente información del archivo Employee Attrition.csv

a) El tiempo de servicio promedio de los trabajadores

```python
seniority = df['Time_of_service'].mean()
print(f"El tiempo de servicio promedio de los trabajadores es de: {seniority:.1f}")

#current dataframe is save to excel
df.to_excel("exercise_a.xlsx")
```

b) El porcentaje de hombres y mujeres en el archivo

```python
#we count the total number of employees, the number of male employees, and the number of female employees to make the calculations
totalEmployees = df['Gender'].count()
maleCount = df[df['Gender'] == 'M']['Gender'].count()
femaleCount = df[df['Gender'] == 'F']['Gender'].count()

print("La cantidad de hombres en el archivo es " + str(maleCount) + " ,equivale al " + "{:.2f}".format((maleCount / totalEmployees) * 100) + "%")
print("La cantidad de mujeres en el archivo es " + str(femaleCount) + " ,equivale al " + "{:.2f}".format((femaleCount / totalEmployees) * 100) + "%")

#current dataframe is save to excel
df.to_excel("exercise_b.xlsx")
```

c) El tiempo máximo sin promoción de las personas que llevan más de 15 años de
servicio

```python
#all values above 15 in the column "time_of_service" are filtered and finally the max value of the column time_since_promotion is calculated
t = df[df['Time_of_service'] > 15]['Time_since_promotion'].max()
print(f"El tiempo máximo sin promoción de las personas con más de 15 años de servicio es de {t}")

#current dataframe is save to excel
df.to_excel("exercise_c.xlsx")
```

d) La edad promedio en la que ingresaron las personas a la empresa, genera una
columna derivada mediante la fórmula: Age – Time_of_service

```python
#a new column named "edad_de_ingreso" is calculated by substracting the "time_of_service" column from the "age" column
df['Edad_de_ingreso'] = df["Age"] - df["Time_of_service"]
#finally the average from the "edad_de_ingreso" column is calculated
edad_ingreso_avg = df["Edad_de_ingreso"].mean()
print(f"La edad promedio de ingreso de las personas es de {edad_ingreso_avg:.1f} años")

#new dataframe is save to excel
df.to_excel("exercise_d.xlsx")
```

e) Un dataframe con las columnas Employee_ID, Relationship_Status y Decision_skill_possess que sólo contenga los registros de mujeres, solteras y con compensación tipo 2 o 3.

```python
#Female employees who are single and with type2 or type3 compensation are filtered
filtered_df = df[(df["Gender"] == "F") &
                 (df["Relationship_Status"] == "Single") &
                 ((df["Compensation_and_Benefits"] == "type2") | (df["Compensation_and_Benefits"] == "type3"))]
#finally the new dataframe is created and printed
result_df = filtered_df[["Employee_ID", "Relationship_Status", "Decision_skill_possess"]]
print(result_df)

#new dataframe is save to excel
result_df.to_excel("exercise_e.xlsx")
```

f) Una columna derivada con el valor “on time” si cuando tuvo su última promoción tenía menos de 30 años. En caso contrario colocar el valor “out of time”

```python
#a new column "last_promotion_age" is calculated by substracting the "time_since_promotion" from "age"
#all values under 30 from the new column created will have "on time" value in a new column called "promotion_time"
#all values greater than and equal to 30 will have "out of time" value  in a new column called "promotion_time"

df["last_promotion_age"] = df["Age"] - df["Time_since_promotion"]
df.loc[df["last_promotion_age"] < 30, "promotion_time"] = "on Time"
df.loc[df["last_promotion_age"] >= 30, "promotion_time"] = "out of time"

#Only printed the following 4 rows to check if dataframe has the correct information:
#Age, Time_since_promotion, last_promotion_age, promotion_time
#and the first 100 rows as well

print(df[["Age", "Time_since_promotion", "last_promotion_age", "promotion_time"]].head(100))

#save to excel the whole dataframe, all rows and all columns
df.to_excel("exercise_f.xlsx")
```
